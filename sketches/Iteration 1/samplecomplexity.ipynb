```python
import numpy as np
import matplotlib.pyplot as plt

# Function to compute sample complexity n
def sample_complexity(epsilon, delta):
    return (1 / (2 * epsilon ** 2)) * np.log(2 / delta)

# Define ranges for epsilon and delta
epsilons = np.linspace(0.005, 0.1, 100)  # Error tolerance
deltas = [0.01, 0.05, 0.1,0.15,0.2]  # Confidence levels (1 - delta)

# Plot sample complexity for different deltas
plt.figure(figsize=(10, 6))
for delta in deltas:
    complexities = sample_complexity(epsilons, delta)
    plt.plot(epsilons, complexities, label=rf'Confidence = $1 - \beta = {1-delta:.2f}$')

# Set plot labels and title with LaTeX code
plt.xlabel(r'Error tolerance ($\epsilon$)')
plt.ylabel(r'Sample complexity ($n$)')
plt.title(r'Sample Complexity for Different Confidence Levels with $\epsilon$ and $\beta$')
plt.legend()
plt.grid(True)
plt.yscale('log')  # Logarithmic scale for better visualization
plt.savefig('sp1.jpg')
plt.show()
```

```python

# Function to compute error tolerance epsilon
def error_tolerance(n, delta):
    return np.sqrt((1 / (2 * n)) * np.log(2 / delta))

# Define range of number of samples
n_samples = np.arange(10, 10000, 10)  # Number of samples

# Define different delta values
deltas =  [0.01, 0.05, 0.1,0.15,0.2]  # Confidence levels (1 - delta)

# Plot error tolerance for different deltas
plt.figure(figsize=(10, 6))
for delta in deltas:
    tolerances = error_tolerance(n_samples, delta)
    plt.plot(n_samples, tolerances, label=rf'Confidence = $1 - \beta = {1-delta:.2f}$')

# Set plot labels and title with LaTeX code
plt.xlabel(r'Number of samples ($n$)')
plt.ylabel(r'Error tolerance ($\epsilon$)')
plt.title(r'Error Tolerance ($\epsilon$) vs. Number of Samples ($n$) for Different $\beta$ Values')
plt.legend()
plt.grid(True)
#plt.yscale('log')  # Logarithmic scale for better visualization
plt.xscale('log')  # Logarithmic scale for better visualization
plt.grid(True)
plt.savefig('sp2.jpg')
plt.show()
```
